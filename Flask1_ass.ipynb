#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a framework for building web applications using the Python programming language. It provides a set of tools and features that make it easier to create websites and web services.

#### Advantages of using Flask:

1.Easy to use: Flask has a simple and straightforward design, making it easy for developers to understand and use. It doesn't overwhelm you with complex features or requirements, so you can start building your web application quickly.

2.Flexible: Flask allows you to choose the tools and components you want to use. It doesn't force you to use specific libraries or technologies, giving you the freedom to build your application the way you want.

3.Small and lightweight: Flask has a small codebase and doesn't require many resources. This means it runs fast and efficiently, making it suitable for small projects or applications with limited resources.

4.Extensible: Flask provides a wide range of extensions that you can add to your application to add extra functionality. These extensions cover things like database integration, user authentication, and more. You can easily pick and choose the extensions that fit your project's needs.

5.Well-documented: 
Flask has excellent documentation that guides you through the process of building web applications. The documentation includes examples and explanations that make it easier for you to understand and use Flask effectively.

#### Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or views in your web application.

When a user requests a specific URL in a Flask application, the app routing mechanism determines which function or view should handle that request and provide a response. This mapping is defined using decorators called @app.route().

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for a specified function or endpoint based on its name and any arguments passed to it.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the homepage!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Profile page of {username}'

@app.route('/product/<int:product_id>')
def product_details(product_id):
    return f'Details of product {product_id}'

if __name__ == '_main_':
    with app.test_request_context():
        home_url = url_for('home')
        user_url = url_for('user_profile', username='john')
        product_url = url_for('product_details', product_id=123)

        print(f'Home URL: {home_url}')
        print(f'User Profile URL: {user_url}')
        print(f'Product Details URL: {product_url}')
